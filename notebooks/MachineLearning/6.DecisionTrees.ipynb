```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
data.keys()
```

```python
# DataFrame으로 만들기
dataset = pd.DataFrame(data=data['data'], columns=data['feature_names'])
dataset
```

```python
# 훈련셋, 테스트셋으로 나누기
from sklearn.model_selection import train_test_split

X = dataset.copy()
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
print(X_train.shape, X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
# 예측결과 클래스별 확률이 나옴
clf.predict_proba(X_test)

# 각 클래스가 0 아니면 1일 확률로 명확히 구분됨
# max_depth가 None이면 Tree가 최대한 자라남
```

```python
# 최대 깊이(Hyper Parameter)를 설정해서 다시 훈련 시작
clf = DecisionTreeClassifier(max_depth=4)
clf = clf.fit(X_train, y_train)
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
# 테스트셋으로 예측한 결과(분류)
```

```python
# 각 클래스별 확률 출력해보자
clf.predict_proba(X_test)
```

```python
# 성능지표
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
# 각 분류별로 몇개씩 출현되었는지
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
# 정답 클래스가 0이며 0으로 예측한 결과가 56번
# 정답 클래스가 1이며 1로 예측한 결과가 102번
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, predictions)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions, target_names=['maligant', 'benign']))
```

```python
# criterion: gini, entropy
# splitter: best, random
# max_features: 특성 중 몇개를 사용할건지(무작위로 선택함)
# random_state: max_features가 최대값보다 작다면 사용할 특성을 랜덤하게 결정
# class_weight: 분류에만 사용, 특정 클래스에 대한 가중치
```
